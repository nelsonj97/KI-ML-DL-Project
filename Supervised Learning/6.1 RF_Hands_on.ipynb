# Random Forest Regression
- random forest regression is very similar to the decision tree regression.<br>
  - The only difference is:
    -  that it uses multiple decision trees to make the final prediction.
    -  the sklearn Class where the model is situated is `RandomForestRegressor` 
       and not DecisionTreeRegressor as it's the case with decision tree.<br>
- All the steps are the same as decision tree regression.<br>
- The only difference is that we will use `RandomForestRegressor` instead of `DecisionTreeRegressor`.<br>
- The rest of the code is exactly the same as decision tree regression.<br>

# Goal:
Is to Train a decision tree model to understand the correlation between Position level and salary
- Reminder: Decision tree is not well adapted for Data sets with only **one features** and one **dependent variables**

# Adaptation Code:
This adaptation code will make the Decision tree to work on any number of features i.e also on<br>
one features.<br>
# Note:
- Feature Scaling is not applied in Decision tree regression and also on random forest.
- This code/implementation works on all the Date set, should just change the name of the data set and select more features<br>
  because here we excluded the first column.

# 0. Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 1. Importing the Data set

```python
# 1. Importing the Data set
data_set = pd.read_csv('Position_Salaries.csv')

# 2. Localizing our features
X = data_set.iloc[:, 1:-1].values

# 3. Localizing our dependent variables
Y = data_set.iloc[:, -1].values
```

# 2. Taking care of missing Value
Here we don't have Missing data

# 3. Encoding Categorical Data
Here we don't have Categorical Variable

# Encoding the Dependent Variables
Here we don't have Categorical Variable

# 4. Training the Decision Tree Regression model on the whole dataset
- We can split the data into training and split set if we want to evaluate our model on 
  new observation. or can just train our model on the whole data<br>

- **Here we want to leverage the maximum Data to understand the correlation in the small amount of information/Data set**.


```python
# 0. Importing the Random Forest Regression Class.
from sklearn.ensemble import RandomForestRegressor

# 1. Building the Random Forest Regression Model
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)    # The most important parameter is n_estimators, which is the...
                                                                          # .... number of trees we want in the forest. i.e 10 trees.
                                                                          # random_state = 0 in oder to get the same result as my friend
# 2. Training the model/regressor on the whole data set
regressor.fit(X, Y)
```

# 5. Feature Scaling
- We should never applied feature scaling on Decision Tree and random forest regression.
  this is because the prediction are resulting from successfully splitting the Data through the different node.

# 6. Predicting a new result
- Here we are going to predict the value 6.5
- Since no feature scaling is needed for we don't have to transform any value

```python
# 1. to predict a new value we have to call the predict() function on our model i.e regressor
regressor.predict([[6.5]])     # Always in double square bracket.
```

Our prediction is bad because it's lower than the requested salary of this employee i.e. 160k $
- Since the model is not well adapted for single value data sets.

# 7. Visualizing the Decision Tree Regression result(higher resolution i.e. 2D)

```python
# For the purpose of having a much more beautiful Tree
# 1. Calling the scatter function from plt module, in order to display the different point/coordinate....
#... containing the result. i.e. 
# plotting salary against Position level
# reverse the scaling
X_grid = np.arange(min(X), max(X), 0.1)        # Since no feature scaling was use   
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, Y, color = 'red' )

# 2. Plotting the SV regression line 
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')  # Applying the predict method on X_grid scale
                                                            # we have to scale x_grid so that predict() method can predict in the right format
plt.title('Truth or Bluff (Random Forest)')
plt.xlabel('Position level')
plt.ylabel('Salary')

# 3. visualizing the Graph by using the show() function
plt.show()
```

- Result is not very pretty because we're just using 1 feature but when using more features it will be very much pretty
- With Random Forest we have more steps than Decision Tree Regression, this is because we're using 10 trees and not just one tree <br>
  as it's the case with Decision Tree Regression.
