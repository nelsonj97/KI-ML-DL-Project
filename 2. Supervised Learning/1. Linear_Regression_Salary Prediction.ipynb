# Linear Regression
Our salary data set contains 30 observation and 2 columns:
- **Year of experience** is our **feature** or **Variable**
- **salary** is our **dependent** variable i.e what we want to predict.

# Goal 
Is to build a simple linear regression model, that will be trained to **understand the correlation between the**<br>
**number of experience and the salary**. so that it will predict for a new employee with a certain number of experience<br>
his corresponding salary.<br>

# 0. Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# 1. Importing the dataset

```python
data_set = pd.read_csv('Salary_Data.csv')
X = data_set.iloc[:, :-1].values
Y = data_set.iloc[:, -1].values
```

# 2. Splitting the data set
Splitting the data set into training set and test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

# 3. Training the Simple Linear Regression
The train set will be used to tain our simple linear regression model

```python
# 1. Import the linear regression model from sklearn 
from sklearn.linear_model import LinearRegression

# 2. We create new variable/object to store our simple linear regression model
regressor = LinearRegression()         # With no parameter, since it will just create a simple linear...
                                       # ... Regression, every simple that we don't need to enter a parameter.

# 3. we used the fit() method to train our model.
regressor.fit(X_train, Y_train)        # the fit() expect a certain format of the training set.
                                       # And this format is the couple of X_train and Y_ train
```

# 4. Evaluating the Model: Predicting the Test set
The Test set will be used to evaluate the model. i.e we are going to...<br>
predict the remaining salary by only using the number of experience. <br>

```python
# 1. we call our predict() method with the help of our regression variable ...
# ... so that it will predict the test set
Y_pred = regressor.predict(X_test)            # Storing the prediction in Y_pred
```

# 5. Visualizing the Training set result
Plotting the real salary compare to the predicted salary

```python

# 1. plotting the scatter plot of the train set
plt.scatter(X_train, Y_train, color = 'red')           # In scatter plot we have to enter the coordinate of X_train and Y_train..
                                                       # ... This coordinates ar nothing but the X_train and the Y_train.
# 2. Plotting the regression line for the training set.
# The regression line is the line of the prediction that comes as close to the real result i.e. to the real salary.
plt.plot(X_train, regressor.predict(X_train), color = 'blue')      # X_train since here we are visualizing the result of the training set.
                                                       # Calling the predict method on X_train, meaning the number of years...
                                                       # .... of experience of the employee, will give us exactly the predicted...
                                                       # .... salary of the training set.
plt.title('Salary Vs Experience (Training Set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()            # in order to display the out put
```

# 6. Visualizing the Test Set Result.

```python
# 1. plotting the scatter plot of the Test set
plt.scatter(X_test, Y_test, color = 'red')             # In scatter plot we have to enter the coordinate of X_train and Y_train..
                                                       # ... This coordinates ar nothing but the X_train and the Y_train.
# 2. Plotting the regression line for the testing set.
# The regression line is the line of the prediction that comes as close to the real result i.e. to the real salary.
# plt.plot(X_test, regressor.predict(X_test), color = 'blue')      # X_test since here we are visualizing the result of the testing set.
                                                       # Calling the predict method on X_test, i.e. the number of years...
                                                       # .... of experience of the employee, will give us exactly the predicted...
                                                       # .... salary of the testing set.
# The regression line is the same for both the training and the test set...
# ... therefor if we replace X_train with X_test in the code below we will still have the same result.
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary Vs Experience (Test Set)')
plt.xlabel('Year of Experience')
plt.ylabel('Salary')
plt.show()            # in order to display the out put
```

The raison why we have a straight blue line/regression line is because they was a perfect linear relationship in
the data set i.e. between the future and the dependent variable of the data set.
